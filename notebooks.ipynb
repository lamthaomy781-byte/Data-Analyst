# Q1: PHÂN TÍCH NHU CẦU & TỐI ƯU CHIẾN LƯỢC NHẬP HÀNG

## 1. Đánh giá chất lượng & quy mô dữ liệu

Tập dữ liệu gồm 3.650 quan sát theo ngày, tương đương hơn 10 năm dữ liệu lịch sử, phản ánh hoạt động nhập và bán cá của nhà hàng trong thời gian dài. Dữ liệu không có giá trị thiếu và các biến định lượng đã được chuẩn hóa về kiểu số, cho thấy quy trình ghi nhận dữ liệu ổn định và nhất quán. Với quy mô và chất lượng này, các thống kê mô tả và phân tích phân phối có độ tin cậy cao, phù hợp để hỗ trợ các quyết định vận hành dài hạn.

## 2. Nhận diện tình trạng thiếu hàng trong vận hành hiện tại

Phân tích biến missed_demand_flag cho thấy 50% số ngày nhà hàng bán hết toàn bộ lượng cá nhập. Điều này đồng nghĩa với việc trong một nửa số ngày, doanh thu của nhà hàng bị giới hạn bởi lượng nhập thay vì phản ánh đầy đủ nhu cầu thực tế của khách hàng. Kết quả này cho thấy tình trạng thiếu cung xảy ra thường xuyên, dù nhu cầu quan sát được không có dấu hiệu biến động mạnh theo thời gian.

## 3. Mối quan hệ giữa lượng nhập và lượng bán

Biểu đồ phân tán giữa stock_units và sold_units cho thấy một tỷ lệ lớn các điểm dữ liệu nằm đúng trên đường sold_units = stock_units, xác nhận rằng trong nhiều ngày toàn bộ lượng cá nhập đều được bán hết. Phần dữ liệu nằm dưới đường chéo phản ánh phân phối nhu cầu thực trong các ngày không bị thiếu hàng, với mức dao động tương đối hẹp.

Từ góc độ kinh doanh, kết quả này cho thấy doanh thu của nhà hàng thường xuyên bị “chặn trần” bởi quyết định nhập hàng, chứ không phải bởi hạn chế từ phía nhu cầu thị trường.

## 4. Ước lượng nhu cầu trung bình hằng ngày

Để tránh sai lệch do dữ liệu bị kiểm duyệt trong các ngày bán hết, nhu cầu trung bình được ước lượng dựa trên các ngày không bị thiếu hàng. Kết quả cho thấy:

- Nhu cầu trung bình ≈ 104.2 đơn vị/ngày

- Nhu cầu trung vị = 105 đơn vị/ngày

- Độ lệch chuẩn ≈ 10 đơn vị

Sự gần nhau giữa giá trị trung bình và trung vị, cùng với độ phân tán vừa phải, cho thấy nhu cầu khách hàng ổn định quanh một mức trung bình và không có dấu hiệu lệch hay biến động bất thường. Điều này xác nhận rằng nhu cầu hằng ngày được ước lượng đáng tin cậy từ dữ liệu lịch sử.

## 5. Đánh giá lợi nhuận theo chiến lược nhập hiện tại

Theo cấu trúc chi phí của đề bài, lợi nhuận trung bình mỗi ngày của chiến lược hiện tại đạt khoảng 210.9 USD. Khi phân tách theo trạng thái thiếu hàng:

- Ngày bán hết hàng có lợi nhuận trung bình cao hơn (≈ 226.7 USD)

- Ngày không bán hết có lợi nhuận trung bình thấp hơn (≈ 195.0 USD)

Mặc dù các ngày bán hết mang lại lợi nhuận cao hơn do không phát sinh chi phí dư hàng, nhà hàng đồng thời mất cơ hội bán thêm trong các ngày này khi nhu cầu thực tế vượt quá lượng nhập.

## 6. Đánh giá tính tối ưu của chiến lược nhập hiện tại

Chiến lược điều chỉnh nhập hiện tại dựa trên việc tăng hoặc giảm ngẫu nhiên 2–5 đơn vị sau mỗi ngày, mang tính phản xạ ngắn hạn. Cách tiếp cận này không sử dụng thông tin về phân phối nhu cầu và không xét đến sự bất cân xứng trong cấu trúc chi phí, khi chi phí thiếu hàng (2 USD/đơn vị) cao hơn chi phí dư hàng (1 USD/đơn vị).

Do đó, dù lợi nhuận quan sát được trong dữ liệu lịch sử ở mức tương đối cao, chiến lược nhập hiện tại chưa được thiết kế để tối đa hóa lợi nhuận kỳ vọng trong dài hạn.

## 7. Chiến lược nhập tối ưu đề xuất và kết quả

Dựa trên cấu trúc chi phí của bài toán, trong đó chi phí thiếu hàng (2 USD/đơn vị) cao hơn chi phí dư hàng (1 USD/đơn vị), mô hình Newsvendor cho thấy mức nhập tối ưu theo lý thuyết nên nằm tại 67% phân vị của phân phối nhu cầu. Áp dụng mô hình này lên phân phối nhu cầu quan sát được (từ các ngày không bị thiếu hàng) cho ra mức nhập tham chiếu khoảng 109 đơn vị/ngày.

Tuy nhiên, do dữ liệu nhu cầu bị kiểm duyệt trong các ngày bán hết, phân phối nhu cầu quan sát được có xu hướng đánh giá thấp nhu cầu thực tế. Để phản ánh rủi ro thiếu hàng đã quan sát được trong vận hành (xảy ra trong 50% số ngày), mức nhập tối ưu được điều chỉnh tăng thêm một độ lệch chuẩn, dẫn đến lượng nhập đề xuất cuối cùng khoảng 119 đơn vị/ngày, cao hơn mức nhập trung bình hiện tại (~115.4 đơn vị/ngày).

Kết quả mô phỏng lợi nhuận dựa trên nhu cầu giả lập cho thấy lợi nhuận trung bình của chiến lược tối ưu sau điều chỉnh đạt khoảng 193.3 USD/ngày, thấp hơn lợi nhuận quan sát được của chiến lược hiện tại (210.9 USD/ngày). Tuy nhiên, chênh lệch này không phản ánh rằng chiến lược tối ưu kém hiệu quả, mà chủ yếu xuất phát từ việc mô phỏng dựa trên phân phối nhu cầu đã bị đánh giá thấp do dữ liệu lịch sử không ghi nhận đầy đủ nhu cầu trong các ngày bán hết.

## Overall

Phân tích dữ liệu 10 năm cho thấy nhà hàng đang vận hành trong bối cảnh nhu cầu khách hàng ổn định nhưng thường xuyên bị thiếu cung, với khoảng 50% số ngày bán hết toàn bộ lượng cá nhập. Điều này dẫn đến việc doanh thu và lợi nhuận quan sát được trong nhiều ngày bị giới hạn bởi quyết định nhập hàng, thay vì phản ánh đầy đủ nhu cầu thị trường.

Chiến lược nhập hiện tại tạo ra lợi nhuận quan sát được tương đối cao trong ngắn hạn, đặc biệt trong các ngày bán hết hàng, nhưng đạt được điều đó bằng cách chấp nhận rủi ro thiếu hàng đáng kể và mất cơ hội bán thêm. Cách điều chỉnh nhập mang tính ngẫu nhiên và phản xạ sau mỗi ngày không tận dụng được thông tin từ dữ liệu lịch sử cũng như sự bất cân xứng trong cấu trúc chi phí thiếu - dư hàng.

Việc áp dụng chiến lược nhập dựa trên dữ liệu, cụ thể là mô hình Newsvendor có điều chỉnh để xử lý nhu cầu bị kiểm duyệt, cho thấy hướng tiếp cận phù hợp hơn trong dài hạn. Dù lợi ích tài chính ngắn hạn chưa thể hiện rõ ràng trong mô phỏng dựa trên dữ liệu lịch sử, chiến lược tối ưu được đề xuất giúp giảm tần suất thiếu hàng, đáp ứng tốt hơn nhu cầu ổn định của khách hàng và tạo nền tảng cho lợi nhuận bền vững và trải nghiệm khách hàng tốt hơn khi nhu cầu thực tế được ghi nhận đầy đủ.
